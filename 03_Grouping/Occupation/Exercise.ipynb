### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep = "|", index_col = "user_id")
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").mean().age
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# converting gender to num
gender_to_num = lambda x: 1 if x=="M" else 0
users["gender_n"] = users["gender"].apply(gender_to_num)
male_ratio_per_occu = users.groupby("occupation").gender_n.sum() / users["occupation"].value_counts() * 100
male_ratio_per_occu.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").age.agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age = users.drop(columns=["gender_n"]).groupby(["occupation","gender"]).mean().rename(columns = {"age":"mean_age" })
mean_age.head()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender_n':'count'})
occup_count = users.groupby(['occupation']).agg({'gender_n':'count'})
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
occup_gender = occup_gender.rename(columns = {"gender_n":"occupation_percentage"})
occup_gender
```
